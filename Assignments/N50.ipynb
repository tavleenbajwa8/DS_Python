```python
import random
import numpy as np
def create_genome(n=10000):
    return ''.join(np.random.choice(list('ACTG'), n))
```

```python
s = "malayalam"
forward = []
reverse = []
for i in range(len(s)-1, -1, -1):
    forward.append(s[i])
for k in range(0, len(s)):
    reverse.append(s[k])
print(forward)
print(reverse)
if forward == reverse:
    print("Palindrome")
else:
    print("Not a Palindrome")
```
