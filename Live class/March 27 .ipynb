```python
df1 = sns.load_dataset('tips')
```

```python
df1
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df1)
```

```python
df1
```

```python
df1['smoker'].value_counts()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df1, hue = 'smoker')  

#hue categorize data using colour by performing groupby operation followed by representing it in different colours
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df1,style = 'smoker') 

#style is same as hue like group by but it use different shape instead of colour
```

```python
df1
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df1,style = 'time', hue = 'smoker') 
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df1, size = 'size')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', kind = 'line', data = df1)
```
