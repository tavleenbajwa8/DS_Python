```python
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
```

```python
df
```

```python
sns.scatterplot(x = df.sepal_length, y = df.petal_length)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df.sepal_length, df.petal_length, 'o')
plt.xlabel("sep_len")
plt.ylabel("plt_len")  #In matplotlib we have to mention x and y axis names
```

```python
sns.distplot(df['sepal_length'])
```

```python
df1 = sns.load_dataset('tips')
```

```python
df1
```

```python
#relplot:relationplot

sns.relplot(x = 'total_bill', y = 'tip', data = df1)
```

```python
df1['smoker'].value_counts()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df1, hue = 'smoker', style= 'smoker')

#hue:Group by operation, group data based on colours
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df1, hue = 'smoker', style= 'size')    
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df1, hue = 'smoker', style= 'time')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df1, size = 'size', style = 'size')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df1, kind = 'line')
```

```python
#Multiple graphs
#Eg: One graph for dinner and other for lunch
```

```python
#Using col
sns.relplot(x = 'total_bill', y = 'tip', data = df1, col= 'time')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df1, col= 'day')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df1, col= 'sex')
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = df1)
```

```python
sns.catplot(x = 'day', y = 'tip', data = df1)
```

```python
sns.catplot(x = 'day', y = 'tip', data = df1, kind = 'swarm')
```

```python
sns.catplot(x = 'day', y = 'tip', kind = 'box', data = df1, hue = 'sex')
```

```python
sns.pairplot(df1)
```
